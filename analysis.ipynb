```python
import torch
import matplotlib.pyplot as plt
import json
```

```python
# Step 3: Load attention weights from the JSON file
with open('/home/duynk29/khanhnnm/checkpoint/CMGAN_unet_32_kd_AFD_full_layer_0.3/checkpoints/attn_weight/attn_weight_1.json', 'r') as f:
    loaded_attention_weights = torch.tensor(json.load(f))

# Step 4: Visualize attention weights
plt.figure(figsize=(10, 5))
plt.imshow(loaded_attention_weights[13].numpy(), cmap='hot', interpolation='nearest')
plt.xlabel('Source tokens')
plt.ylabel('Target tokens')
plt.title('Attention weights')
plt.colorbar()
plt.show()
```
